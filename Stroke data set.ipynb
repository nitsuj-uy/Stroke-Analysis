```python
import pandas as pd
import random
import sys
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import statsmodels.api as sm
np.set_printoptions(threshold=sys.maxsize)
```

```python
stroke = pd.read_csv('healthcare-dataset-stroke-data.csv')
```

```python
print(list(stroke))
print(stroke.dtypes)
#print(stroke.head(5))
```

```python
stroke['gender'] = stroke['gender'].replace(['Other'], 'Male')
```

Categorical variables are gender, hypertension, heart_disease, ever_married, work_type, Residence_type, smoking_status

Categorical variables to convert to dummy : gender, ever_married, work_type, Residence_type, smoking_status

```python
stroke['smoking_status'] = stroke['smoking_status'].astype('category')
stroke['smoking_status'] = stroke['smoking_status'].cat.reorder_categories(['never smoked', 'formerly smoked', 'smokes', 
                                                            'Unknown'], ordered = True)
stroke['smoking_status'] = stroke['smoking_status'].cat.codes
print(stroke['smoking_status'])
```

0 - never smoked    1-formerly smoked    2-smokes    3- unknown

```python
stroke['work_type'] = stroke['work_type'].astype('category')
stroke['work_type'] = stroke['work_type'].cat.reorder_categories(['Private', 'Self-employed', 'Govt_job', 'children',
                                                                 'Never_worked'], ordered = True)
stroke['work_type'] = stroke['work_type'].cat.codes
print(stroke['work_type'])
```

0-private 1-self-employed 2-Govt_job 3-children 4-Never_worked

https://analyticsindiamag.com/a-hands-on-guide-to-dummy-variable-trap-with-a-solution-in-python/

```python
stroke = pd.get_dummies(stroke, columns = ['gender', 'ever_married', 'Residence_type'], drop_first = True)
```

1-Male, 0 Female, 1 Married, 0 Not married, 1- urban, 2-rural

```python
y = stroke['stroke']
x = stroke.drop(columns = ['stroke','id'])
```

```python
stroke_logit_model = sm.Logit(y,x)
result = stroke_logit_model.fit()
print(result.summary2())
```

The p-value for gender_Male is greater than 0.5. This predictor can be dropped to improve the logsitic model. 

```python
x = x.drop(columns = ['gender_Male'])
```

```python
stroke_logit_model = sm.Logit(y,x)
result = stroke_logit_model.fit()
print(result.summary2())
```

```python
stroke_logit_model = sm.Logit(y,x)
result = stroke_logit_model.fit()
print(result.summary())
```

Test the Logistic Regression Model

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 4)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
LR = LinearRegression()
LR.fit(x_train, y_train)
```

```python
LR.predict(x_test)
```

```python
LR.score(x_test,y_test)
```

```python
LogR = LogisticRegression()
LogR.fit(x_train, y_train)
```

```python
print(LogR.coef_, LogR.intercept_)
```

```python
log_prediction = LogR.predict(x_test)
print(classification_report(y_test, log_prediction))
```

```python
probabilities = LogR.predict_proba(x_test)
print(probabilities)
print(log_prediction)
```

```python
confusion_matrix(y_test, log_prediction)
```

True Negatives upper left

False Negatives lower left

False positive upper right

True positives lower right

```python
LogR.score(x_test,y_test)
```
